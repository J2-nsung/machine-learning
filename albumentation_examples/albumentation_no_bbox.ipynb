```python
import random

import cv2
from matplotlib import pyplot as plt

import albumentations as A

%matplotlib inline
```

```python
def visualize(image):
    plt.figure(figsize=(8, 8))
    plt.axis('off')
    plt.imshow(image)
```

```python
img = cv2.imread(f'./data/VOC2007/JPEGImages/000001.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
visualize(img)

img0 = cv2.imread(f'./data/VOC2007/JPEGImages/000019.jpg')
img0 = cv2.cvtColor(img0, cv2.COLOR_BGR2RGB)
visualize(img0)

img1 = cv2.imread(f'./data/VOC2007/JPEGImages/000028.jpg')
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
visualize(img1)
```

```python
transform = A.Compose(
    [A.VerticalFlip(p=1)], # albumentations.VerticalFlip : 상하 반전
    additional_targets={'image0': 'image', 'image1': 'image'}
)
```

```python
transformed = transform(image=img, image0=img0, image1=img1)
```

```python
visualize(transformed['image'])
```

```python
visualize(transformed['image0'])
```

```python
visualize(transformed['image1'])
```

```python
# 좌우 반전 and rotate and 밝기,대비 조절
transform = A.Compose(
    [
        A.HorizontalFlip(p=0.5),
        A.ShiftScaleRotate(p=0.5),
        A.RandomBrightnessContrast(p=0.2),
        A.RGBShift(p=0.2),
    ],
    additional_targets={'image0': 'image', 'image1': 'image'}
)
```

```python
random.seed(42) # ??
transformed = transform(image=img, image0=img0, image1=img1)
```
